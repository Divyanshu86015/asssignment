Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer =  There are five plots  that we can plot using the Seaborn libreary.

1.Scatter Plot (scatterplot):

Use: To visualize the relationship between two continuous variables. It helps in identifying the correlation, trend, and distribution of the data points.
Example: Examining the relationship between height and weight.

2.Histogram (histplot):

Use: To display the distribution of a single continuous variable. It helps in understanding the frequency distribution, skewness, and the presence of outliers.
Example: Analyzing the distribution of ages in a population.

3.Line Plot (lineplot):

Use: To display data points in a time series or to show trends over time. It's useful for showing how a variable changes over a continuous interval or time.
Example: Plotting stock prices over time to observe trends.

4.Bar Plot (barplot):

Use: To show the relationship between a categorical variable and a continuous variable. It is used for comparing the sizes of different groups or categories.
Example: Comparing average scores of students in different subjects.

5.Box Plot (boxplot):

Use: To show the distribution of a continuous variable by displaying the data quartiles and averages. It is useful for detecting outliers and understanding the spread and skewness of the data.
Example: Comparing the distribution of salaries across different departments in a company.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset("fmri")
sns.lineplot(x="timepoint" , y="signal",data=df,hue="event", style="region")
plt.title("FMRI")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event/Region")
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
df = sns.load_dataset("diamonds")
sns.histplot(x=df.price,data=df,hue="cut")
plt.title("DIMONDS")
plt.xlabel("Price")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
