Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Answer =  Glyphs are the visual representations of data points or elements in a plot. They are essentially the shapes and symbols used to depict data in a Bokeh plot. Common glyphs include points (e.g., circles, squares), lines, bars, and more complex shapes like patches and wedges. Each glyph has a set of properties that can be customized, such as size, color, and opacity.italicised text

Common Glyphs in Bokeh:

Scatter: Represents data points .

Line: Connects data points with lines.

Bar: Creates bar charts.

Patch: Draws polygons.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Various Glyphs Example")
p.xaxis.axis_label = "paltal lenght"
p.yaxis.axis_label = "paltal widhth"

# Add a scatter plot with circles
p.scatter(x, y, size=10, color="navy", alpha=0.5, legend_label="Circles")

# Add a line plot
p.line(x, y, line_width=2, color="green", legend_label="Line")

# Add a bar plot
p.vbar(x=x, top=y, width=0.5, color="orange", legend_label="Bars")

# Show the plot
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]

output_file("line.html")

line1 = p.line(x, y, legend_label="Line 1", line_width=2, line_color="red")
p.title.text = "Customized Bokeh Plot"
p.title.align = "center"
p.title.text_font_size = "15pt"
p.title.text_color = "#501240"

p.xaxis.axis_label = "X-Axis"
p.xaxis.axis_line_width = 2
p.xaxis.axis_line_color = "blue"
p.yaxis.axis_label = "Y-Axis"
p.yaxis.axis_line_width = 2
p.yaxis.axis_line_color = "green"

p.legend.title = 'Legend'
p.legend.location = 'top_left'
p.legend.background_fill_color = '#b6b60d'

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

 ANswer = A Bokeh server allows us to create interactive plots that can be updated in real time. It enables us to build and deploy interactive web applications that connect front-end user interactions to Python code running on a server. This is particularly useful for applications requiring dynamic updates, real-time data visualization, and user interactions.


Key Concepts

Bokeh Document: Represents all the content of the Bokeh application (plots, widgets, etc.). It's the central object that Bokeh uses to manage updates and interactions.

Callbacks: Functions that are triggered by events, such as user interactions (e.g., button clicks, slider changes). Callbacks allow us to update plots and widgets in response to user input.
