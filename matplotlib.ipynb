Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Answer = A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It is particularly useful for identifying patterns, trends, correlations, and outliers within the data. In a scatter plot, each data point is represented as a dot on a two-dimensional plane, where the x-axis represents one variable and the y-axis represents the other variable.

```python
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

Answer = It  provides a way to plot multiple plots on a single figure. Given the number of rows and columns , it returns a tuple ( fig , axa ), giving a single figure fig with an array of axes axa .

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2,2)
axs[0, 0].plot(x1, y1, label='Line1', color= '#D106A3')
axs[0, 0].set_title('Line plot1')
axs[0, 0].legend()

axs[0, 1].plot(x2, y2, label='Line 2', color='#D1A306')
axs[0, 1].set_title('Line plot2')
axs[0, 1].legend()

axs[1, 0].plot(x3, y3, label='Line 3', color='#06D122')
axs[1, 0].set_title('Line plot3')
axs[1, 0].legend()

axs[1, 1].plot(x4, y4, label='Line 4', color='#06C2D1')
axs[1, 1].set_title('Line plot 4')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Answer = A bar plot is a graphical representation of data using rectangular bars to show the frequency, count, or any other measure for different categories. The length or height of the bars represents the value of the measure.

Why used ?
Comparison: Bar plots are useful for comparing different categories or groups.

Clarity: They make it easy to see and understand the differences between the values of the categories.

Versatility: They can be used for both discrete and categorical data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(company,profit,color =['#7A2F01',"green","#D3E9F9","orange",])
plt.xlabel("profit")
plt.ylabel("company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Answer = A box plot (also known as a whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It can also show outliers and variability outside the upper and lower quartiles.

Why Used? ---Summary of Data: It provides a visual summary of one or more sets of data, highlighting the median, quartiles, and potential outliers

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Distributions')
plt.show()
```
