Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
predicting the quality of wine.


Answer = Here are the key features:

1. Fixed Acidity: Represents non-volatile acids such as tartaric acid that do not evaporate easily. Acidity contributes to the tartness and freshness of wine.

2. Volatile Acidity: Refers to the amount of acetic acid in the wine. High volatile acidity is associated with spoilage, giving the wine a vinegar-like taste.

3. Citric Acid: A weak organic acid that adds freshness to wine. It prevents spoilage and balances the taste.

4. Residual Sugar: The amount of sugar left after fermentation. This can influence the sweetness of the wine, but balanced residual sugar levels help improve taste.

5. Chlorides: The salt content in wine. Higher levels of chlorides can give a salty taste, which can negatively impact wine quality.

6. Free Sulfur Dioxide (SO2): This component prevents microbial growth and oxidation. It's crucial for preserving the wine, but too much can affect its flavor.

7. Total Sulfur Dioxide: The sum of free and bound sulfur dioxide. Like free SO2, it helps preserve the wine but must be kept at balanced levels.

8. Density: Wine density, related to sugar content and alcohol, can impact wine's body and texture. It correlates with residual sugar and alcohol content.

9. pH: A measure of the wine's acidity or alkalinity. Lower pH (higher acidity) often leads to fresher, more balanced wine.

10. Sulphates: Act as preservatives and antioxidants. Sulphates help stabilize wine, but too much can affect taste.

11. Alcohol: The alcohol content by volume (ABV). Alcohol significantly influences the flavor, body, and warmth of the wine. Balanced alcohol is associated with higher wine quality.

12. Wine Quality (Target): The target variable representing the quality of wine on a scale from 0 to 10, based on sensory evaluations (taste, aroma, etc.).


# Importance of Features in Predicting Wine Quality:

- Volatile Acidity and Alcohol are often found to be key predictors of wine quality. High volatile acidity is associated with spoilage, while alcohol content significantly impacts the wine's overall balance and perception.

- Sulphates, citric acid, and free sulfur dioxide help in maintaining the stability and freshness of the wine, which correlates positively with quality ratings.

- pH and fixed acidity also play essential roles as acidity is critical in determining the wine's balance and taste, contributing to its overall score.

- Residual Sugar and chlorides have subtler impacts but still contribute to the taste and texture of the wine.



Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.

Answer =1. Simple Imputation Techniques

1.  Mean/Median Imputation

 - How it works: Replace missing values with the mean (or median) of the column.
 - When to use: Suitable for numeric features that are symmetrically distributed (mean) or skewed (median).

 **Advantages:**

 - Simple to implement.

 - Works well for small amounts of missing data (especially if less than 5% of the data is missing).

 **Disadvantages:**

 - Reduces variance: This can distort the distribution of the data, making it less representative of reality.

 - Can introduce bias: Particularly if the missing values are not missing completely at random (e.g., if higher quality wines have systematically missing values).

1. Mode Imputation

  - How it works: Replace missing values with the most frequent value (mode) for categorical variables.

 - When to use: Suitable for categorical variables with relatively few unique values.

 **Advantages:**

 - Works well for categorical features where one category is dominant.

- **Disadvantages:**

 - Can skew the dataset: It overrepresents the most frequent category.

 - Ignores relationships between features, which may introduce bias.

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

```python
import pandas as pd
df = pd.read_csv('winequality-red.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = df.drop('quality', axis=1)
X_scaled = StandardScaler().fit_transform(X)
X_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
```

```python
import numpy as np
min_components = np.argmax(cumulative_variance >= 0.90) + 1
min_components
```
